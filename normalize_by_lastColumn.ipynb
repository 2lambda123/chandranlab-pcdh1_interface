```python
import pandas as pd
from os import listdir
from os.path import isfile, join
```

Getting the list of files to be processed

```python
# competition
# indir = '/Users/gorkalasso/Dropbox (EinsteinMed)/Elisa_platform/megan_competition_elisas/PCDH1 Mapping_ Competition ELISAs of EC1 mutants/'
# indir = '/Users/gorkalasso/Dropbox (EinsteinMed)/Elisa_platform/megan_pcdh1_gngc/10_27_21/PCDH1 Mapping_ Competition ELISAs of EC1 mutants/'
# neutralization
# indir = '/Users/gorkalasso/Dropbox (EinsteinMed)/Elisa_platform/megan_pcdh1_gngc/PCDH1 Mapping_ rVSV-ANDV Neutralization with EC1 mutants/'
indir = '/Users/gorkalasso/Dropbox (EinsteinMed)/Elisa_platform/megan_pcdh1_gngc/10_27_21/PCDH1 Mapping_ rVSV-ANDV Neutralization with EC1 mutants/'
filelist = []
for f in listdir(indir):
    if not isfile(join(indir, f)):
        continue
    if (not '.xlsx' in f) or (f[0:1] == '~'):
        continue
#     print(f)
    filelist.append(f)
print('Files to normalize: ', len(filelist))
```

Normalize each file and concatenate them

```python
df = pd.DataFrame()
for c, f in enumerate(filelist):
#     if c > 0:
#         break
    print(c, f)
    # Loading excel into dataframe
    df0 = pd.concat(pd.read_excel(indir + f, sheet_name=None), ignore_index=True)
    # Renaming columns
    df0 = df0.rename(columns={'Unnamed: 0': 'Instance'})
    for col in df0.columns:
#         print(col)
        if isinstance(col, float):
#             print('\tfloat')
            df0 = df0.rename(columns={col: "{:.6f}".format(col)})
#     display(df0)
    # Normalizing
    df1 = df0.copy()
    df1.iloc[:,1:] = df0.iloc[:,1:].div(df0['Control'], axis=0)*100
#     display(df1)
    # Adding to main dataframe
    df1['File'] = f
    df = pd.concat([df, df1])
# df.to_excel('competition_elisa_normalized.xlsx')
df.to_excel('neutralization_normalized.xlsx')
```
